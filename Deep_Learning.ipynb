```python
dataFrame = pd.read_csv('/content/drive/MyDrive/csvler/librosasonmfcc40/7li.csv')
selRows = dataFrame[dataFrame['Duygu'] == 'human2'].index
dataFrame = dataFrame.drop(selRows, axis=0)

dataFrame = dataFrame.drop("Unnamed: 0", axis=1)
dataFrame = dataFrame.drop("Unnamed: 0.1", axis=1)
dataFrame= dataFrame.replace(to_replace ="oksurme",value =3)
dataFrame= dataFrame.replace(to_replace ="ciglik",value =4)
dataFrame= dataFrame.replace(to_replace ="nefes",value =5)

dataFrame= dataFrame.replace(to_replace ="human",value =0)
dataFrame= dataFrame.replace(to_replace ="dogs",value =1)
dataFrame= dataFrame.replace(to_replace ="cats",value =2)
```

```python
y = dataFrame["Duygu"].values
X = dataFrame.drop("Duygu",axis = 1).values
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=15)
scaler = MinMaxScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
model = Sequential()
model.add(layers.Dense(256, activation='relu', input_shape=(x_train.shape[1],)))
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=50,batch_size=128)
action = model.predict(x_test)
```
